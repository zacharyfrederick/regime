# Feature reduction

SHAP, VIF, or correlation clustering on master_features to reduce multicollinearity and select features for the regime model. See guide section 8.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import config
import pandas as pd

if config.MASTER_FEATURES_PATH.exists():
    df = pd.read_parquet(config.MASTER_FEATURES_PATH)
    numeric = df.select_dtypes(include=["number"]).columns.tolist()
    print("Numeric columns:", len(numeric))
    # Correlation matrix for clustering
    corr = df[numeric].corr()
    print(corr.shape)
else:
    print("Run pipeline first.")
```
