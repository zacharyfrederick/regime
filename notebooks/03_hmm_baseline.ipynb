# HMM baseline

Load master feature table, select feature columns, and fit a baseline regime model (e.g. hmmlearn). Filter by date range and optionally sector/cap tier.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import config
import pandas as pd

if config.MASTER_FEATURES_PATH.exists():
    df = pd.read_parquet(config.MASTER_FEATURES_PATH)
    # Filter to date range from config
    df = df[(df["date"] >= config.DATE_START) & (df["date"] <= config.DATE_END)]
    print("Loaded master, shape:", df.shape)
    # Select numeric feature columns for HMM (exclude identity/metadata)
    feature_cols = [c for c in df.columns if c not in ("ticker", "date", "sector", "famaindustry") and df[c].dtype in ("float64", "float32")]
    print("Feature columns:", len(feature_cols))
else:
    print("Run pipeline first.")
```
